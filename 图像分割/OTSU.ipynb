```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline

img = Image.open('Lenna.png')
out = img.convert("L")
im = np.array(out)
var_max = th_max = 0

for n in range(255):
    fro=bac=mfro=mefro=pfro=var=0
    mean = np.mean(im)
    size = im.size
    for i in np.nditer(im, op_flags = ['readwrite']):
        if i < n:
            mfro+=i
            fro+=1
    try:
        mefro = mfro/fro
        pfro = fro/size
        pfro = pfro/(1-pfro)
        var = pfro*((mefro-mean)**2)
    except ZeroDivisionError:
        continue
    if var>var_max:
        var_max = var
        th_max = n

print('maximum variance is',var_max, 'maximum threshold is',th_max)
aa=np.where(im>th_max,True,False)
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(aa,cmap='gray')
```

```python
img.close()
otsuimg = Image.fromarray(aa)
```
