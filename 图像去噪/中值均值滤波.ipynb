```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline

img = Image.open('D:\JupyterFile\Lena_p&snoise.png')
```

```python
def median_filter(img,k=3):
    rgb = np.array(img)
    for n in range(3):
        bb = np.pad(rgb[:,:,n],int((k-1)/2),'edge')
        for i in range(img.size[0]):
            for j in range(img.size[1]):
                rgb[i,j,n] = np.median(bb[i:i+k,j:j+k])
    return Image.fromarray(rgb)

def mean_filter(img,k=3):
    rgb = np.array(img)
    for n in range(3):
        bb = np.pad(rgb[:,:,n],int((k-1)/2),'edge')
        for i in range(img.size[0]):
            for j in range(img.size[1]):
                rgb[i,j,n] = int(np.mean(bb[i:i+k,j:j+k]))
    return Image.fromarray(rgb)
```

```python
imgmedf3 = median_filter(img,3)
imgmedf5 = median_filter(img,5)
imgmeaf3 = mean_filter(img,3)
imgmeaf5 = mean_filter(img,5)
```

```python
plt.subplot(2,3,1)
plt.imshow(img)
plt.subplot(2,3,2)
plt.imshow(imgmedf3)
plt.subplot(2,3,3)
plt.imshow(imgmedf5)
plt.subplot(2,3,4)
plt.imshow(img)
plt.subplot(2,3,5)
plt.imshow(imgmeaf3)
plt.subplot(2,3,6)
plt.imshow(imgmeaf5)
```
