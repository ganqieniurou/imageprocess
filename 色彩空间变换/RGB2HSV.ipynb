```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline

img = Image.open('D:\JupyterFile\Lenna.png')

#RGB2HSV
rgb = np.array(img) / 255
rgb1 = np.array(img) 
vmax = np.max(rgb,axis=2)
vmin = np.min(rgb,axis=2)
vdif = vmax - vmin

maskz = vdif == 0 
maskr = vmax == rgb[:,:,0]
maskg = vmax == rgb[:,:,1]
maskb = vmax == rgb[:,:,2]
masks = vmax == 0
vmaxnz = vmax + masks
vdifnz = vdif + maskz

h = np.float64(maskz)#cla formu: https://www.rapidtables.com/convert/color/rgb-to-hsv.html
h += maskr * ((60 * ((rgb[:,:,1] - rgb[:,:,2]) / vdifnz)) + 360)%360
h += maskg * ((60 * ((rgb[:,:,2] - rgb[:,:,0]) / vdifnz)) + 120)%360
h += maskg * ((60 * ((rgb[:,:,0] - rgb[:,:,1]) / vdifnz)) + 240)%360
h = np.uint16(np.rint(h))
s = ~masks * (vdif/vmaxnz)
v = vmax
# s = np.uint8(~masks * (vdif/vmaxnz) * 255)
# v = np.uint8(vmax * 255)
hsv = np.transpose(np.array([h,s,v]),(1,2,0))

#HSV2RGB
c = hsv[:,:,1] * hsv[:,:,2] #cla formu: https://www.rapidtables.com/convert/color/hsv-to-rgb.html
x = c * (1 - abs(hsv[:,:,0] / 60 % 2 - 1))
m = hsv[:,:,2] - c
frgb = np.zeros([img.size[0],img.size[0],3])

maskang = np.uint8(hsv[:,:,0] / 60)
frgb[:,:,0] = (maskang == 0) * c
frgb[:,:,1] = (maskang == 0) * x
frgb[:,:,1] += (maskang == 1) * c
frgb[:,:,0] += (maskang == 1) * x
frgb[:,:,1] += (maskang == 2) * c
frgb[:,:,2] += (maskang == 2) * x
frgb[:,:,2] += (maskang == 3) * c
frgb[:,:,1] += (maskang == 3) * x
frgb[:,:,2] += (maskang == 4) * c
frgb[:,:,0] += (maskang == 4) * x
frgb[:,:,0] += (maskang == 5) * c
frgb[:,:,2] += (maskang == 5) * x
frgb = np.uint8(np.rint((frgb + np.transpose([m,m,m],(1,2,0)))*255))

plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(np.uint8(frgb))
```
