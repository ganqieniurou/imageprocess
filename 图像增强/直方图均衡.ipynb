```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline

img = Image.open('D:\JupyterFile\Lenna.png')

def histogram_equal(img):
    rgb = np.array(img)
    for n in range(3):
#         his = np.histogram(rgb[:,:,n].flatten(),range(257))[0]/(img.size[0]*img.size[1])
        hist = np.uint8((np.histogram(rgb[:,:,n].flatten(),range(257))[0]/(img.size[0]*img.size[1])).cumsum()*255)
        for i in range(img.size[1]):
            for j in range(img.size[0]):
                rgb[i,j,n] = hist[rgb[i,j,n]]
    return Image.fromarray(rgb)
```

```python
plt.subplot(121),plt.imshow(img)
plt.subplot(122),plt.imshow(histogram_equal(img))
```
