```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline

img = Image.open('D:\JupyterFile\Lenna.png')
```

```python
#Laplace Operator
def laplace_operator(img,k=1):
    laplace_filter = np.array([[-1,-1,-1],[-1, 8,-1],[-1,-1,-1],])
    rgb = np.array(img)
    for n in range(3):
        bb = np.pad(rgb[:,:,n],1,'edge')
        for i in range(img.size[0]):
            for j in range(img.size[1]):
                if np.sum(bb[i:i + 3,j:j + 3] * laplace_filter) > 50:
                    rgb[i,j,n] += int(k * np.sum(bb[i:i + 3,j:j + 3] * laplace_filter))
    return Image.fromarray(rgb)

```

```python
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(laplace_operator(img,0.2))
```
